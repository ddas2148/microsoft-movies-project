```python
import gzip
import os
import shutil
zip_dir = 'zippedData'
out_dir = 'data'
zip_files = [f for f in os.listdir(zip_dir) if f.endswith('.gz')]

for zip_file in zip_files:
    zip_path = os.path.join(zip_dir, zip_file)
    out_path = os.path.join(out_dir, zip_file[:-3])
    with gzip.open(zip_path, 'rb') as f_in:
        with open(out_path, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
```
